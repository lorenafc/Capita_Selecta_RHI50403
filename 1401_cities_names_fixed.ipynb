```python
import pandas as pd
```

```python
# Load the spreadsheets
authors_cities_not_geocoded = pd.read_excel('authors_cities_not_geocoded.xlsx')
authors_cities_not_geocoded_done = pd.read_excel('authors_cities_not_geocoded_done.xlsx')
```

```python

# Ensure both DataFrames are sorted by 'indexauthor'
authors_cities_not_geocoded = authors_cities_not_geocoded.sort_values(by='indexauthor').reset_index(drop=True)
authors_cities_not_geocoded_done = authors_cities_not_geocoded_done.sort_values(by='indexauthor').reset_index(drop=True)

# Add 'indexauthor' column to both DataFrames
authors_cities_not_geocoded['indexauthor'] = authors_cities_not_geocoded['indexauthor']
authors_cities_not_geocoded_done['indexauthor'] = authors_cities_not_geocoded_done['indexauthor']

# Find the differences
differences = authors_cities_not_geocoded_done.compare(authors_cities_not_geocoded)

# Reset the index of differences DataFrame
differences_with_index = differences.reset_index()

# Ensure 'indexauthor' is included
# Use the original DataFrame to add the 'indexauthor' column
differences_with_index['indexauthor'] = authors_cities_not_geocoded_done['indexauthor']

# Adjust the column order
# Extract the multi-level column names
multi_index_columns = differences_with_index.columns
flat_columns = [f'{col[0]}_{col[1]}' if col[1] != '' else col[0] for col in multi_index_columns]

# Create a mapping from flat column names to multi-index names
col_mapping = {flat: multi_index_columns[i] for i, flat in enumerate(flat_columns)}

# Create a DataFrame with flat columns to facilitate column reordering
differences_with_index.columns = flat_columns

# Reorder columns to include 'indexauthor' at the beginning
if 'indexauthor' in differences_with_index.columns:
    column_order = ['indexauthor'] + [col for col in flat_columns if col != 'indexauthor']
    differences_with_index = differences_with_index[column_order]



```

```python
differences_with_index = differences_with_index.drop('index', axis=1)
```

```python

# Define the columns to be renamed
columns_to_rename = {
    'borncity_self': 'borncity_fixed',
    'deathcity_self': 'deathcity_fixed',
    'activecity_self': 'activecity_fixed',
    'longitude_born_self': 'longitude_born_fixed',
    'latitude_death_self': 'latitude_death_fixed',
    'longitude_active_self': 'longitude_active_fixed'
}

# Replace 'self' with 'fixed' in the column names
renamed_columns = {
    col: columns_to_rename.get(col, col)  # Get new name if in rename dictionary, else keep old name
    for col in differences_with_index.columns
}

# Rename the columns
differences_with_index.rename(columns=renamed_columns, inplace=True)

# Reorder columns to include 'indexauthor' at the beginning
if 'indexauthor' in differences_with_index.columns:
    column_order = ['indexauthor'] + [col for col in renamed_columns.values() if col != 'indexauthor']
    differences_with_index = differences_with_index[column_order]


```
