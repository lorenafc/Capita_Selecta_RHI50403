```python
file_name = 'merged_additional_authors_encoded.xlsx'
authors_small_cities = pd.read_excel(file_name, engine='openpyxl')
authors_small_cities_copy=authors_small_cities.copy()
```

```python
# create an index number for each author
authors_small_cities.rename(columns={'Unnamed: 0': 'indexauthor'}, inplace=True)
```

```python
# Filter deathyear from 800 to 1800
authors_small_cities_800_1800 = authors_small_cities[
    (authors_small_cities['deathyear'] >= 800) & (authors_small_cities['deathyear'] <= 1800)
]
```

```python
authors_small_cities_800_1800.head()
```

```python
cities_not_geocoded = authors_small_cities_800_1800[authors_small_cities_800_1800[['latitude_death', 'longitude_death', 'latitude_active', 'longitude_active']].isna().all(axis=1)]
```

```python
cities_not_geocoded.head()
```

```python
total_800_1800 = len(authors_small_cities_800_1800)
non_geocoded_800_1800 = len(cities_not_geocoded)

# Calculate the rate of authors with not geocoded cities from 800 to 1800
percentage_not_geocoded = non_geocoded_800_1800 / total_800_1800

print("The total of authors", total_800_1800)
print("The total of authors with active and death cities not geocoded", non_geocoded_800_1800)
print("The percentage of authors with active and death cities not geocoded from 8000 to 1800 is:", round(percentage_not_geocoded*100, 2),"%")
```

```python
# export the result as excel file
file_name = 'authors_cities_not_geocoded_800_1800.xlsx'
cities_not_geocoded.to_excel(file_name)
```
