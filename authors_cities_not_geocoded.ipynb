# HOW MANY AUTHORS WITH SOME PLACE NOT GEOCODED

```python

# Create conditions where 'borncity', 'deathcity', or 'activecity' is not null and their corresponding latitudes and longitudes are NaN
condition_born = authors_small_cities['borncity'].notna() & (authors_small_cities['latitude_born'].isna() | authors_small_cities['longitude_born'].isna())
condition_death = authors_small_cities['deathcity'].notna() & (authors_small_cities['latitude_death'].isna() | authors_small_cities['longitude_death'].isna())
condition_active = authors_small_cities['activecity'].notna() & (authors_small_cities['latitude_active'].isna() | authors_small_cities['longitude_active'].isna())


# Combine the conditions
condition = condition_born | condition_death | condition_active

# Use this condition to filter your DataFrame
authors_small_cities_NaN_born_death_active = authors_small_cities[condition]
```

```python
authors_small_cities_NaN_born_death_active.head()
```

```python
authors_small_cities_NaN_born_death_active.shape
```

```python
total = len(authors_small_cities)
non_geocoded = len(authors_small_cities_NaN_born_death_active)

# Calculate the rate of authors with not geocoded cities
percentage_not_geocoded = non_geocoded / total

print("The total of authors is", total)
print("The total of authors with cities not geocoded is", non_geocoded)
print("The percentage of authors with cities not geocoded is:", round(percentage_not_geocoded*100, 2),"%")
```

```python
# export the result as excel file
file_name = 'authors_cities_not_geocoded.xlsx'
authors_small_cities_NaN_born_death_active.to_excel(file_name)
```

## HOW MANY PLACES NOT GEOCODED

```python
# Combine all places from borncity, deathcity, and activecity into a single list
all_places = pd.concat([
    authors_small_cities['borncity'].dropna(),
    authors_small_cities['deathcity'].dropna(),
    authors_small_cities['activecity'].dropna()
]).unique()

# Convert the combined list to a set to ensure all places are unique
unique_total_places = set(all_places)

# Now do the same for non-geocoded places
not_geocoded_places = pd.concat([
    authors_small_cities[condition_born_not_geocoded]['borncity'].dropna(),
    authors_small_cities[condition_death_not_geocoded]['deathcity'].dropna(),
    authors_small_cities[condition_active_not_geocoded]['activecity'].dropna()
]).unique()

# Convert to a set to ensure uniqueness
unique_not_geocoded_places = set(not_geocoded_places)

# Count the number of unique places and unique non-geocoded places
total_unique_places = len(unique_total_places)
total_unique_not_geocoded_places = len(unique_not_geocoded_places)

# Calculate the percentage of unique places that were not geocoded
percentage_unique_not_geocoded = total_unique_not_geocoded_places / total_unique_places * 100

print("The total number of unique places is:", total_unique_places)
print("The total number of unique places not geocoded is:", total_unique_not_geocoded_places)
print("The percentage of unique places not geocoded is:", round(percentage_unique_not_geocoded, 2), "%")

# Export the data to an Excel file
file_name = 'unique_places_not_geocoded.xlsx'
authors_small_cities[condition_born_not_geocoded | condition_death_not_geocoded | condition_active_not_geocoded].drop_duplicates().to_excel(file_name)
```
